Q1. Solve the anagram using brute foce technique

PAY * MY = MONEY

```python
from itertools import permutations

# Function to convert letter digits to number
def to_number(letters, digits):
    return int(''.join(str(digits[letter]) for letter in letters))

# Set of unique letters in PAY * MY = MONEY
unique_letters = 'PAYMONE'

# Store solutions in a list
solutions = []

# Check all permutations of 10 digits taken 7 at a time (for 7 unique letters)
for perm in permutations(range(10), len(unique_letters)):
    digits = dict(zip(unique_letters, perm))
    
    PAY = to_number('PAY', digits)
    MY = to_number('MY', digits)
    MONEY = to_number('MONEY', digits)
    
    if PAY * MY == MONEY:
        solutions.append((digits['P'], digits['A'], digits['Y'], digits['M'], digits['O'], digits['N'], digits['E'], PAY, MY, MONEY))

# Print all solutions
for solution in solutions:
    P, A, Y, M, O, N, E, PAY, MY, MONEY = solution
    print(f"P={P}, A={A}, Y={Y}, M={M}, O={O}, N={N}, E={E}")
    print(f"PAY = {PAY}, MY = {MY}, MONEY = {MONEY}\n")
```

Q2. Write a word count program in Python and display the count of words in descending order of their frequencies.

```python
import re

def count_words(text):
    # Normalize text to lowercase and remove punctuation
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    
    # Split the text into words
    words = text.split()
    
    # Count the frequency of each word using a dictionary
    word_counts = {}
    for word in words:
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1
    
    return word_counts

def display_word_counts(word_counts):
    # Sort the words by frequency in descending order
    sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
    
    # Display the results
    print(f"{'Word':<15} | {'Count':<5}")
    print('-' * 22)
    for word, count in sorted_word_counts:
        print(f"{word:<15} | {count:<5}")

# Sample text for testing
text = """
Python is an interpreted, high-level, general-purpose programming language. Python's design philosophy emphasizes code readability with the use of significant indentation. 
Its language constructs and object-oriented approach aim to help programmers write clear, logical code for small and large-scale projects.
"""

# Count words
word_counts = count_words(text)

# Display word counts
display_word_counts(word_counts)
```
