Q1. Follow along with these steps:
a) Create a figure object called fig using plt.figure()
b) Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
c) Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
import matplotlib.pyplot as plt

# Create a figure object
fig = plt.figure()

# Add an axis to the figure
ax = fig.add_axes([0, 0, 1, 1])

# Plot y = 2x
x = range(0, 100)  # x values from 0 to 100
y = [2 * xi for xi in x]  # y = 2x
ax.plot(x, y, label='y = 2x')

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Plot of y = 2x')
ax.legend()

# Set x-ticks at intervals of 20
ax.set_xticks(range(0, 101, 20))

# Set y-ticks at intervals of 50
ax.set_yticks(range(0, 201, 50))

# Show the plot
plt.show()
```

Q2. Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2]
respectively. Now plot (x,y) on both axes. And call your figure object to show it.

```python
import matplotlib.pyplot as plt

# Create a figure object
fig = plt.figure()

# Add two axes to the figure
ax1 = fig.add_axes([0, 0, 1, 1])       # Full figure (Position [0,0,1,1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])  # Smaller inset (Position [0.2,0.5,0.2,0.2])

# Data to plot
x = range(0, 101)  # x values from 0 to 100
y = [2 * xi for xi in x]  # y = 2x

# Plot (x, y) on both axes
ax1.plot(x, y, label='y = 2x')
ax2.plot(x, y, label='y = 2x')

# Set labels and titles for both axes
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Full Plot of y = 2x')
ax1.legend()

ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Inset Plot of y = 2x')
ax2.legend()

# Display the figure
plt.show()
```

Total profit data provided for each month. Generated line plot must include the following properties: –

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('company_sales_data.csv')

# Extract the Total Profit for each month
months = df['month_number']
total_profit = df['total_profit']

# Create the line plot
plt.figure(figsize=(10, 6))  # Optional: Adjust figure size for better readability
plt.plot(months, total_profit, marker='o', linestyle='-', color='b', label='Total Profit')

# Set labels and title
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit of All Months')
plt.legend()

# Show the plot
plt.grid(True)  # Optional: Add grid for better readability
plt.show()
```

Q4. Use the company sales dataset csv file, get total profit of all months and show line plot with the following
Style properties. Generated line plot must include following Style properties: –
a. Line Style dotted and Line-color should be red
b. Show legend at the lower right location.
c. X label name = Month Number
d. Y label name = Sold units number
e. Add a circle marker.
f. Line marker color as read
g. Line width should be 3

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('company_sales_data.csv')

# Extract the Total Profit and Month Number
months = df['month_number']
total_profit = df['total_profit']

# Create the line plot with the specified style properties
plt.figure(figsize=(10, 6))  # Optional: Adjust figure size for better readability
plt.plot(months, total_profit, 
         linestyle='--',  # Dotted line style
         color='red',     # Line color
         marker='o',      # Circle marker
         markerfacecolor='red',  # Marker color
         markeredgecolor='red',  # Marker edge color
         linewidth=3,     # Line width
         label='Total Profit')  # Label for the legend

# Set labels and title
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.title('Total Profit of All Months')

# Show legend at the lower right location
plt.legend(loc='lower right')

# Optional: Add grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```

Q1. Use the company sales dataset csv file, read all product sales data and show it using a multiline plot.
Display the number of units sold per month or each product using multiline plots. (i.e., Separate Plotline
for each product).

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('company_sales_data.csv')

# Extract the columns: 'month_number' and the product columns for the plot
product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
sales_data = df[['month_number'] + product_columns]

# Create the multiline plot
plt.figure(figsize=(12, 8))  # Optional: Adjust figure size for better readability

# Plot each product's sales data
for product in product_columns:
    plt.plot(sales_data['month_number'], sales_data[product], marker='o', label=product)

# Set labels and title
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Number of Units Sold Per Month for Each Product')

# Show legend
plt.legend(title='Product')

# Show grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```

Q2. Use the company sales dataset csv file, calculate total sale data for last year for each product and show it
using a Pie chart.
Note: In Pie chart display Number of units sold per year for each product in percentage.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('company_sales_data.csv')

# Filter the data for the last year
last_year_data = df[df['month_number'] > df['month_number'].max() - 12]

# Calculate the total units sold for each product in the last year
product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
total_units_last_year = last_year_data[product_columns].sum()

# Create a pie chart showing the percentage of units sold for each product
plt.figure(figsize=(10, 7))  # Optional: Adjust figure size for better readability
plt.pie(total_units_last_year, 
        labels=total_units_last_year.index, # uses the product names as labels.
        autopct='%1.1f%%', # displays the percentage of each slice
        startangle=140, # rotates the pie chart for better readability
        colors=plt.cm.Paired(range(len(total_units_last_year)))) # provides distinct colors for each segment.

# Set title
plt.title('Percentage of Units Sold Per Product for the Last Year')

# Display the pie chart
plt.show()
```
