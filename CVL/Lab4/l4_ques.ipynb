```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_sobel_operator(image_path):
    # Load the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Define the Sobel kernels
    sobel_x = np.array([[ -1, 0, 1],
                        [ -2, 0, 2],
                        [ -1, 0, 1]], dtype=np.float32)
    
    sobel_y = np.array([[ -1, -2, -1],
                        [  0, 0,  0],
                        [  1, 2,  1]], dtype=np.float32)
    
    # Apply the Sobel operator
    grad_x = cv2.filter2D(image, cv2.CV_64F, sobel_x)
    grad_y = cv2.filter2D(image, cv2.CV_64F, sobel_y)
    
    # Compute the magnitude of the gradient
    magnitude = np.sqrt(grad_x**2 + grad_y**2)
    
    # Normalize the result to fit in the 0-255 range
    magnitude = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX)
    
    # Convert to uint8
    magnitude = np.uint8(magnitude)
    
    # Display the result
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(image, cmap='gray')
    
    plt.subplot(1, 2, 2)
    plt.title('Sobel Operator')
    plt.imshow(magnitude, cmap='gray')
    
    plt.show()

# Replace 'image.jpg' with your image path
apply_sobel_operator('valve.png')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_laplacian_filter(image_path):
    # Load the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Define the Laplacian kernel
    laplacian_kernel = np.array([[0,  1, 0],
                                [1, -4, 1],
                                [0,  1, 0]], dtype=np.float32)
    
    # Apply the Laplacian filter
    laplacian = cv2.filter2D(image, cv2.CV_64F, laplacian_kernel)
    
    # Normalize the result to fit in the 0-255 range
    laplacian = cv2.normalize(laplacian, None, 0, 255, cv2.NORM_MINMAX)
    
    # Convert to uint8
    laplacian = np.uint8(laplacian)
    
    # Display the result
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(image, cmap='gray')
    
    plt.subplot(1, 2, 2)
    plt.title('Laplacian Filter')
    plt.imshow(laplacian, cmap='gray')
    
    plt.show()

# Replace 'image.jpg' with your image path
apply_laplacian_filter('astroflag.jpg')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_unsharp_masking(image_path, sigma=1.0, alpha=1.5):
    # Load the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Apply Gaussian blur
    blurred = cv2.GaussianBlur(image, (0, 0), sigma)
    
    # Compute the sharpened image
    sharpened = cv2.addWeighted(image, 1.0 + alpha, blurred, -alpha, 0)
    
    # Display the result
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(image, cmap='gray')
    
    plt.subplot(1, 2, 2)
    plt.title('Unsharp Masking')
    plt.imshow(sharpened, cmap='gray')
    
    plt.show()

# Replace 'image.jpg' with your image path
apply_unsharp_masking('usm-img.jpg')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def obtain_gradient(image_path):
    # Load the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Define Sobel kernels
    sobel_x = np.array([[ -1, 0, 1],
                        [ -2, 0, 2],
                        [ -1, 0, 1]], dtype=np.float32)
    
    sobel_y = np.array([[ -1, -2, -1],
                        [  0, 0,  0],
                        [  1, 2,  1]], dtype=np.float32)
    
    # Compute gradients
    grad_x = cv2.filter2D(image, cv2.CV_64F, sobel_x)
    grad_y = cv2.filter2D(image, cv2.CV_64F, sobel_y)
    
    # Compute gradient magnitude
    magnitude = np.sqrt(grad_x**2 + grad_y**2)
    
    # Normalize the result
    magnitude = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX)
    magnitude = np.uint8(magnitude)
    
    # Display the result
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(image, cmap='gray')
    
    plt.subplot(1, 2, 2)
    plt.title('Gradient Magnitude')
    plt.imshow(magnitude, cmap='gray')
    
    plt.show()

# Replace 'image.jpg' with your image path
obtain_gradient('astroflag.jpg')
```
